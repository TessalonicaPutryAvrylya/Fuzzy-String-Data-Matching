```python
from fuzzywuzzy import fuzz
import pandas as pd
import numpy as np
import fuzzywuzzy
from fuzzywuzzy import process
import chardet

# membaca data csv menggunakan pandas
df = pd.read_csv("/content/netflix_titles.csv")
# membangkitkan angka random
np.random.seed(0)
```

```python
# memasukan kolom title kedalam list
film = df['title']
film.values.tolist()
```

```python
# mengambil nama film yang unik (tidak ada duplikat) data dataframe
film = df['title'].unique()
# melakukan sorting
film.sort()

inputu = input("Film yang mau dicari: ")
cocok = process.extract(inputu, film, limit=5, scorer=fuzz.WRatio)
cocok
```
